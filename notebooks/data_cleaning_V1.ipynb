## Missing Values Strategy
We distinguish three types of missing values:

1. Structural missing → kept (e.g. no repair history)
2. Informative missing → flagged (e.g. AGE)
3. Random missing → treated case by case

We avoid global imputation to preserve signal and prevent bias.

MERGING DATASETS

```python
# ==============================
# 0. IMPORTS + PATH
# ==============================

from pathlib import Path

OUTPUT_PATH = Path("output/tables")
OUTPUT_PATH.mkdir(parents=True, exist_ok=True)
```

```python
# ==============================
# 1. BASE TABLE (AGGREGATED)
# ==============================

df_final = df_agg.copy()

print("Base shape:", df_final.shape)
```

```python
# ==============================
# 2. MERGE CLIENTS
# ==============================

cols_clients = [
    "CLIENT_ID",
    "AGE",
    "AGE_KNOWN",
    "GENDER",
    "COUNTRY_OF_RESIDENCY",
    "MonoMultiMarket"
]

df_final = df_final.merge(
    df_clients[cols_clients],
    on="CLIENT_ID",
    how="left"
)

print("After Clients merge:", df_final.shape)
```

```python
# ==============================
# 3. TRANSACTIONS AGGREGATION
# ==============================

df_trs_agg = df_trs.groupby("CLIENT_ID").agg(
    TRS_TOTAL_SPEND=("TO_WITHOUTTAX_EUR_CONST", "sum"),
    TRS_N_TRANSACTIONS=("TO_WITHOUTTAX_EUR_CONST", "count"),
    TRS_TOTAL_QTY=("QTY_PDT", "sum"),
    TRS_AVG_SPEND=("TO_WITHOUTTAX_EUR_CONST", "mean")
).reset_index()
```

```python
df_trs_agg.head()
df_trs_agg.shape
```

```python
# ==============================
# 4. MERGE TRANSACTIONS FEATURES
# ==============================

df_final = df_final.merge(
    df_trs_agg,
    on="CLIENT_ID",
    how="left"
)

print("After Transactions merge:", df_final.shape)
```

```python
df_final[[
    "TRS_TOTAL_SPEND",
    "TRS_N_TRANSACTIONS"
]].describe()
```

```python
# ==============================
# 5. CRC AGGREGATION
# ==============================

df_crc_agg = df_crc.groupby("CLIENT_ID").agg(
    CRC_N_INTERACTIONS=("CREATION_DATE", "count"),
    CRC_LAST_CONTACT=("CREATION_DATE", "max")
).reset_index()
```

```python
df_crc_agg.head()
df_crc_agg.shape
```

```python
# ==============================
# 6. MERGE CRC FEATURES
# ==============================

df_final = df_final.merge(
    df_crc_agg,
    on="CLIENT_ID",
    how="left"
)

print("After CRC merge:", df_final.shape)
```

```python
# ==============================
# 7. HANDLE CRC MISSING
# ==============================

df_final["CRC_N_INTERACTIONS"] = df_final["CRC_N_INTERACTIONS"].fillna(0)
```

```python
# ==============================
# 8. CCP AGGREGATION
# ==============================

df_ccp_agg = df_ccp.groupby("CLIENT_ID").agg(
    CCP_N_PRODUCTS=("ARTICLE_ID", "count"),
    CCP_N_GIFTS=("FLAG_GIFT", "sum"),
    CCP_N_SALES_KNOWN=("SALE_DATE_KNOWN", "sum")
).reset_index()
```

```python
df_ccp_agg.shape
```

```python
# ==============================
# 9. MERGE CCP FEATURES
# ==============================

df_final = df_final.merge(
    df_ccp_agg,
    on="CLIENT_ID",
    how="left"
)

print("After CCP merge:", df_final.shape)
```

```python
df_final = df_final.merge(
    df_ccp_agg,
    on="CLIENT_ID",
    how="left"
)

print("After CCP merge:", df_final.shape)
```

## Merging Pipeline — Overview

The final modeling dataset is built using a structured and scalable pipeline.

### Base Table
- The Aggregated dataset is used as the backbone
- Each row represents a `(CLIENT_ID, DATE_TARGET)` pair
- It already contains target variables and historical aggregated features

---

### Step 1 — Merge Client Information
- Added demographic and static features:
  - AGE, AGE_KNOWN
  - GENDER
  - COUNTRY_OF_RESIDENCY
  - MonoMultiMarket
- Merge key: `CLIENT_ID`
- Join type: left join

---

### Step 2 — Transactions Aggregation
- Transactions are aggregated at the client level before merging
- Features created:
  - TRS_TOTAL_SPEND
  - TRS_N_TRANSACTIONS
  - TRS_TOTAL_QTY
  - TRS_AVG_SPEND
- Rationale: avoid row explosion and ensure scalability
- Merge key: `CLIENT_ID`

---

### Step 3 — CRC (Customer Interactions)
- Aggregated interaction data:
  - CRC_N_INTERACTIONS
  - CRC_LAST_CONTACT
- Missing values handled:
  - No interaction → count = 0
- Merge key: `CLIENT_ID`

---

### Step 4 — CCP (Product Ownership)
- Aggregated ownership features:
  - CCP_N_PRODUCTS
  - CCP_N_GIFTS
  - CCP_N_SALES_KNOWN
- Missing values handled:
  - No ownership → values set to 0
- Merge key: `CLIENT_ID`

---

### Key Design Principles

- Aggregation is performed before merging to avoid data duplication
- All merges are performed using left joins to preserve the base structure
- Missing values are handled explicitly and often used as signals
- Feature names are prefixed (TRS_, CRC_, CCP_) for traceability
